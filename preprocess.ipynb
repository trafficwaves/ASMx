```python
import pandas as pd
data = pd.read_csv('data/2025-04-09.csv')
```

```python
# make the time_unix_fix to find its cloest 30 seonds
data['time_unix'] = data['time_unix_fix'].copy()
data['time_unix_fix'] = pd.to_datetime(data['time_unix'], unit='s')
data['time_unix_fix'] = data['time_unix_fix'].dt.floor('30S')
data['time_unix_fix'] = data['time_unix_fix'].astype('int64') // 10**9
```

```python
# data drop the Unnamed: 0 column
data = data.drop(columns=['Unnamed: 0'])
data.to_csv('data/2025-04-09.csv', index=False)
```
