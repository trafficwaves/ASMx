```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
from ASM_utils import AdaptiveSmoothing
from matplotlib.colors import LinearSegmentedColormap
import warnings 
import os
import random
import time
warnings.filterwarnings("ignore")
device = torch.device("cpu")
lane = 1
rds = np.load('../data/I95speed.npy')
rds_filled = pd.DataFrame(rds).ffill(axis=0).bfill(axis=0).ffill(axis=1).values
speed = np.load('../data/I95speed.npy')
speed[speed < 0] = np.nan
# clip speed to a max of 80mph
speed = np.clip(speed, 0, 80)
space_size, time_size = speed.shape
speed = torch.tensor(speed, dtype=torch.float32).unsqueeze(0).unsqueeze(0)  # Add batch and channel dimensions
best_model_path = f'../logs/calibration/20250607_203336/best_model_lane{lane}.pt'

dx = 0.02                  # distance per cell
dt = 4.0                    # time per cell
kernel_time_window = time_size * dt  # seconds
kernel_space_window = space_size * dx  # same units as dx
```

```python
# load the best model
best_model_path = f'../model/20250607_221107/best_model_lane{lane}.pt'
calibrated_model = torch.load(best_model_path, map_location=device)
print(calibrated_model.keys())
calibrated = AdaptiveSmoothing(kernel_time_window,
                          kernel_space_window,
                          dx, dt,
                          init_tau= calibrated_model['tau'],
                          init_delta= 2*calibrated_model['delta'],
                          init_c_cong=-calibrated_model['c_cong'],
                          init_c_free=-calibrated_model['c_free'],
                          init_v_delta= calibrated_model['v_delta'],
                          init_v_thr= calibrated_model['v_thr'],
                          ).to(device)

calibrated.eval()
```

```python
with torch.no_grad():
    start_time = time.time()
    smoothed = calibrated(speed)
    end_time = time.time()
    print(f"Computation took {end_time - start_time:.2f} seconds")
    calibrated_asm_speed = smoothed.cpu().numpy().squeeze()
```

```python
plt.figure(figsize=(30, 12))
plt.rcParams.update({'font.size': 35, 'font.family': 'serif'})
plt.imshow(calibrated_asm_speed, cmap='RdYlGn', interpolation='nearest', origin='lower',vmin=0, vmax=80, aspect='auto')
plt.colorbar(label='Speed')
plt.title('Interstate 95 Corriror reconstruction with ASM')
plt.tight_layout()
# reverse y-axis to match the original plot
# plt.gca().invert_yaxis()
import os
if not os.path.exists('../figures/figure10'):
    os.makedirs('../figures/figure10')
plt.savefig(f'../figures/figure10/VT_lane{lane}.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
