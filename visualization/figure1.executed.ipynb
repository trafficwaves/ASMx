```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Palatino']
from matplotlib.colors import LinearSegmentedColormap
MILE_TO_KM = 1.60934
```

```python
lane = 1
motion = np.load(f'../data/processed_data/motion/lane{lane}/2024-07-09.npy')
rds = np.load(f'../data/processed_data/rds/lane{lane}/2024-07-09.npy')
rds_filled = pd.DataFrame(rds).bfill(axis=0).ffill(axis=0).ffill(axis=1).values
```

```python
plt.figure(figsize=(24, 6))
plt.rcParams.update({'font.size': 30, 'font.family': 'serif'})
plt.imshow(MILE_TO_KM *motion, cmap='RdYlGn', interpolation='nearest', origin='lower',vmin=0, vmax=120, aspect='auto')
plt.colorbar(label='Speed (km/h)')
plt.title('Ground truth')
plt.tight_layout()
# reverse y-axis to match the original plot
plt.gca().invert_yaxis()
plt.savefig(f'../figures/figure1/MOTION_lane{lane}.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(24, 6))
plt.rcParams.update({'font.size': 30, 'font.family': 'serif'})
plt.imshow(MILE_TO_KM * rds, cmap='RdYlGn', interpolation='nearest', origin='lower',
           vmin=0, vmax=120, aspect='auto')
plt.colorbar(label='Speed (km/h)')
plt.title('RDS raw data')
plt.tight_layout()
# reverse y-axis to match the original plot
plt.gca().invert_yaxis()
plt.savefig(f'../figures/figure1//RDS_raw_lane{lane}.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(24, 6))
plt.rcParams.update({'font.size': 30, 'font.family': 'serif'})
plt.imshow(MILE_TO_KM * rds_filled, cmap='RdYlGn', interpolation='nearest', origin='lower',
           vmin=0, vmax=120, aspect='auto')
plt.colorbar(label='Speed (km/h)')
plt.title('Simple filled RDS')
plt.tight_layout()
# reverse y-axis to match the original plot
plt.gca().invert_yaxis()
plt.savefig(f'../figures/figure1//RDS_fill_lane{lane}.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
